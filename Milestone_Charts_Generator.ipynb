```
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()    
objects = ('Baseline', 'Random Synonym Insertion', 'Synonym Replacement', 'Synonym Appended', '3-layers Encoder-Decoder', '4-layers Encoder-Decoder', '6-layers Encoder-Decoder')
y_pos = np.arange(len(objects))
performance = [ 91.8689, 91.5662, 89.9189, 89.7952, 91.7722, 91.6056, 91.1866]

plt.barh(y_pos, performance, align='center', alpha=0.5)

for i, v in enumerate(performance):
    ax.text(v + 1, i, str(v), color='blue', fontweight='bold')

plt.yticks(y_pos, objects)
plt.xlabel('Validation Accuracy')
plt.title('Validation accuracy for various experiments (English Gloss)')

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

t = ('Baseline\n(2-layers)\n (EN)', 'Random \nSynonym Insertion\n(EN)', 'Synonym Replacement \n(EN)', 'Synonym Appended \n(EN)', 
     '3-layers \nEncoder-Decoder\n (EN)', '4-layers \nEncoder-Decoder\n (EN)', '6-layers \nEncoder-Decoder\n (EN)',
     'Baseline (DE)', 'Random Swap (DE)', 'Random Deletion (DE)')
data1 = [ 91.8689, 91.5662, 89.9189, 89.7952, 91.7722, 91.6056, 91.1866, 55, 35, 37]
data2 = [ 536, 554, 543, 544, 620, 704, 872, 561, 240, 240]
df = pd.DataFrame(list(zip(t, data1, data2)), 
               columns =['Model', 'Validation Accuracy (Bleu Score)', 'Model Size']) 

ax = df.plot(title="Experiments Comparison",
            kind= 'bar',
            x = 'Model',
            # xlabel = 'Experimental Models',
            secondary_y= ['Model Size'], 
            # rot=0,
            # table = True,
            figsize = (10,5),
            legend = False,
            color=["#58508d", '#ff6361'],
        )

# for p in ax.patches:
#     ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
ax1, ax2 = plt.gcf().get_axes() # gets the current figure and then the axes
for p in ax1.patches:
  ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
for p in ax2.patches:
  ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 0.005))
  
ax1.set_ylabel('Validation Accuracy (Bleu Score)', color = '#58508d')
ax2.set_ylabel('Model Size', color ='#ff6361')
# plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()
```
